## <font color=blue> Calculus - Euler Forward Method for Gravitational Acceleration

This is part of the series of math algorithms to define 3D shapes as math objects.



### <font color=blue> Problem: 
Drop object. Calculate distance traversed and velocity at time intervals.


## <font color=purple> SECTION 1: Gravitational Acceleration

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### <font color=blue>Process

```python
# set the values of the t, x, and v
# arrays to implement the Forward 
# Euler Method for num_steps many steps.


def forward_euler():
    """
    Breaks into small increments
    Returns 3 lists (time, distance, and velocity)
    """
    h = 0.1                                   # seconds
    g = 9.81                                  # gavitational acceleration m/sec**2

    num_steps = 50

    t = np.zeros(num_steps + 1)            # time
    x = np.zeros(num_steps + 1)            # distance
    v = np.zeros(num_steps + 1)            # velocity

    for step in range(num_steps):
        t[step + 1] = t[step] + h             # increment time
        x[step + 1] = x[step] +  h * v[step]  # increment distance
        v[step + 1] = v[step] - h * g         # increment velocity
    return t, x, v

t, x, v = forward_euler()


def plot_data():
    plt.style.use('ggplot')
    axes_height = plt.subplot(211)
    plt.subplots_adjust( bottom=-1)
    plt.plot(t, x)
    axes_velocity = plt.subplot(212)
    plt.plot(t, v)
    axes_height.set_ylabel('Height in meter')
    axes_velocity.set_ylabel('Velocity in meters/second')
    axes_velocity.set_xlabel('Time in seconds')
    plt.show() 

plot_data()

acceleration = (v[50] - v[0]) / (t[50] - t[0])
print('Gravitational Acceleration: {:.3f} meter / second**2'.format(acceleration))
```
